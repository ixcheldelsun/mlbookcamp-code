```python
lotus_mask = data['Make'] == 'Lotus'

lotus_cars = data[lotus_mask]
lotus_cars = lotus_cars[["Engine HP", "Engine Cylinders"]]
lotus_cars = lotus_cars.drop_duplicates()
X = lotus_cars.to_numpy()

XTX = np.dot(X.T, X)
XTX = XTX.astype('int64')
#XTX_inv = np.invert(XTX)
y = np.array([1100, 800, 750, 850, 1300, 1000, 1000, 1300, 800])
```

```python
aux = np.dot(np.linalg.inv(XTX), X.T)
w = np.dot(aux, y)

print(w)
```
